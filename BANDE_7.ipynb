```r
tic()
for(i in length(file_list)){
#OBJECT RASTER 
rst <- raster(file_list[i])
mylayer <- stack(rst)
p <- buffer(file_shp, 20)
mylayers.crop <- crop(mylayer, p) 
mylayers.mask <- mask(mylayers.crop, p)
mylayers.point <- as.data.frame(rasterToPoints(mylayers.mask))
d <- dim(mylayers.point)
d1 <- d[1] 
d2 <- d[2]
}
toc()
mylayers.point
```

```r
#ESTRAGGO DATI TELERIVELAMENTO
dati <- mylayers.point[,3:d2] #valori pixel
dati.coord <- mylayers.point[,1:2] #valori coord dei pixel
dati <-as.numeric(dati)
dati <-ceiling(dati)
dati <- order(dati)
dati.order <- dati #ordine decrescente
dati.ordered.t <- as.data.frame(t(dati.order))
```

```r

doy.sep <- colsplit(as.vector(rownames(dati.ordered.t)), "\\.", c(1:2))
dati.ordered.t$DOY <- as.numeric(doy.sep[,1])
dati.ordered.t$lambda <- lambda[i]
library(data.table)
dati.doy <- merge.data.table(data.frame(DOY = 1:365), dati.ordered.t, all.x = TRUE)
dati2fit <-data.frame()
dati2fit <- rbind(dati2fit, dati.doy)
```

```r
dati2fit
```

```r
#ERRORE PERCHE NON PUO INTEROPOLARE VALORI NULL
df.pred <- expand.grid(lambda = seq(490,2130, by = 35),DOY=seq(4,365, by = 7))
DOY <- dati2fit$DOY
lambda <- dati2fit$lambda
f1 <- function(x) { return(predict(gam(as.vector(x/10000) ~ s(lambda,k = c(8), bs = 'cr') + s(DOY, k = c(12), bs = 'cp'), 
                                        na.action = na.exclude, betar(link = "logit"), method = "REML"),df.pred, type = 'response')) }

tic()
dati.tempi.fit <- apply(dati2fit[,2:ncol(dati2fit)-1], MARGIN = 2, f1)
dati.tempi.fit <- dati.tempi.fit[,2:ncol(dati.tempi.fit)]#tolgo la prima colonna
toc()
```

```r
#4.Costruire dataframe ben definito
dati.tempi.fit
punti <- ncol(dati.tempi.fit)
bande <- length(seq(490,2130, by = 35))
tempi <- length(seq(4,365, by = 7))


#5.Costruisco l'array (nxm)con immagini bimdimensionali vuote
myArray <- array(0, dim=c(punti,bande,tempi))
for(i in 1:ncol(dati.tempi.fit)){
    myArray[i,,] <- t(as.matrix(dati.tempi.fit[,i]))
 
```

```r
#6.Creo nuova funzione di classe funData 
g1 <- new("funData", 
     argvals = list( seq(490,2130, by = 35),
     1:tempi),
     X = myArray)

multi_data <-multiFunData(g1)
#7.Definisco componenti principali multidimensionali
tic()
pca <- MFPCA(multi_data,M=35,
       uniExpansions = list(list(type = "splines2D", k = c(8,12))))

toc() 
```

```r
dir.create("Dati")
file.create("MPCA_conero.rda")
save (pca, file = "MPCA_conero.rda")
save (dati.coord,file = "MPCA_conero_coord.rda")
pca$score
dim(pca$score)
head(pca$score)

#7.Associo componenti con coordinate 
pca.comp <- rasterXYZ(cbind(pca$score,dati.coord))
#8.Visualizza le singole ocurve FPCA 
plot(pca.comp[7]) #immagine finale
plot(pca$score[1],obs=1,main="FPCA1")
plot(pca$score[1],obs=2,main="FPCA2") 

#9.Vilualizza curve Mfpca 
n<-20 
m<-20
for(i in 1:n) {
     for(j in 1:m) {
        windows() 
        par(mfrow=c(5,4))
        plot(pca$score[i],obs=j,main = "MFPCA")
}
i+=1
}  
```
