```r
#TIMES SERIES
ts_path<-"TS_ndvi.txt"
ts <- t(read.delim(ts_path,header = T, sep = " ", quote = "\"", dec = "."))
ts
```

```r
#DATAFRAME
tp<- dim(ts)
tp1 <- tp[1]
tp2 <- tp[2]

dati<- as.vector(as.matrix((ts)))
time <- as.numeric(rep(1:tp1, tp2))
ID <- rep(1:tp2, each = tp1)
df_ts <- as.data.frame(cbind(ID, dati, time))
```
